```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Beispieldaten
data = np. array ([2 , 4, 7, 10, 15, 20])

# Plot der ursprungliche Daten
plt.plot(data, marker='o', color='r', label="ursprüngliche Daten")
plt.title('Ursprungliche Daten')

# Min -Max - Normalisierung
min_val = min(data)
max_val = max(data)
normalized_data = ( data - min_val) / ( max_val - min_val)

# Plot der normalisierten Daten
plt.plot(normalized_data, marker='o', color='b', label="Normalisierte Daten")
plt.title("Normalisierte Daten (MinMax)")
plt.legend()
plt.show()
```

### Mache es für zwei-dimensionale Daten mit der Standardisierung

```python
# Zufällige Daten generieren
np.random.seed(42)

# Zufällige Daten für Dimension 1 mit Normalverteilung generieren
data_dim1 = np.random.normal(loc=2, scale=0.2, size=(100, 1))

# Zufällige Daten für Dimension 2 mit Normalverteilung generieren
data_dim2 = np.random.normal(loc=2, scale=2, size=(100, 1))

# data[:, 1].std()  --> 2
# data[:, 0].std() --> 0.2

# Die beiden Dimensionen zusammenführen
data = np.hstack((data_dim1, data_dim2))
```

```python
df = pd.DataFrame(data, columns = ['Dim1', 'Dim2'])
df.head()
```

```python
# np.std(data, axis=0) --> array([0.18072324, 1.89777728]) around the scale value
# np.mean(data, axis=0)  --> array([1.9792307 , 2.04460917])   around the loc value
```

```python
# Funktion zur Standardisierung der Daten pro Dimension
def standardize_data(data):
    mean = np.mean(data, axis=0)
    std_dev = np.std(data, axis=0)
    standardized_data = (data - mean) / std_dev
    return standardized_data

standardized_data = standardize_data(data)
```

```python
# Ausgabe der Mittelwerte und Standardabweichungen pro Dimension
mean_per_dimension = np.mean(data, axis=0)
std_dev_per_dimension = np.std(data, axis=0)
print("Mittelwerte pro Dimension: ", mean_per_dimension)
print("Standardabweichungen pro Dimension: ", std_dev_per_dimension)
```

```python
# Streuudiagram (Scatterplot) der ursprünglichen Daten
plt.figure(figsize=(10,7))
plt.subplot(1, 2, 1)
plt.scatter(data[:, 0], data[:, 1])
plt.title("Scatterplot der unsprünglichen Daten")
plt.xlabel("Dimension 1")
plt.ylabel("Dimension 2")
plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.grid(True)

# Streuudiagram (Scatterplot) der standardisierten Daten
plt.subplot(1, 2, 2)
plt.scatter(standardized_data[:, 0] , standardized_data[:, 1])
plt.title("Scatterplot der standardisierten Daten")
plt.xlabel("Dimension 1")
plt.ylabel("Dimension 2")
plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.grid(True)

plt.tight_layout()
plt.show()
```
